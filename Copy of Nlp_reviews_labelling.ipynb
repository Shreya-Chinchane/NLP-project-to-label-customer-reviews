*** 2) Code for Sentiment analysis by doing basic nlp steps, like stemmimg, lemmatization, stopword removal ***

```python
from google.colab import files
uploaded = files.upload()
```

```python
import docx2txt

# Read the reviews from the file
text = docx2txt.process("reviews.docx")

# Split the reviews into individual sentences
sentences = text.split('.')
```

```python
import nltk
nltk.download('omw-1.4')
nltk.download('vader_lexicon')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

# Download the necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Define the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Define the stop words
stop_words = set(stopwords.words('english'))

# Define a function to get the WordNet POS tag for a given tag
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Preprocess the reviews
preprocessed_reviews = []
for sentence in sentences:
    # Tokenize the sentence
    words = word_tokenize(sentence.lower())
    # Remove stop words
    words = [word for word in words if word not in stop_words]
    # Perform stemming and lemmatization
    stemmed_words = [stemmer.stem(lemmatizer.lemmatize(word, get_wordnet_pos(pos_tag)))
                     for word, pos_tag in nltk.pos_tag(words)]
    # Join the stemmed words into a sentence
    preprocessed_sentence = ' '.join(stemmed_words)
    preprocessed_reviews.append(preprocessed_sentence)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Classify the reviews
for review in preprocessed_reviews:
    # Get the sentiment scores
    sentiment_scores = sia.polarity_scores(review)
    # Classify the review as positive, negative, or neutral based on the compound score
    if sentiment_scores['compound'] >= 0.05:
        print(f"Positive review: {review}")
    elif sentiment_scores['compound'] <= -0.05:
        print(f"Negative review: {review}")
    else:
        print(f"Neutral review: {review}")
```
